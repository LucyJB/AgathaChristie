```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AgathaChristieRawData.csv')
```

```python
# Replace '1' with 'Male' and '0' with 'Female' in the 'gender' column
df['Gender'] = df['Gender'].replace({1: 'Male', 0: 'Female'})
```

```python
# Display the gender split of murderers
murderers_gender_split = df[df['Murderer'] == 1]['Gender'].value_counts()

print("Gender Split of Murderers:")
print(murderers_gender_split)
```

```python
# Plotting
plt.figure(figsize=(8, 6))
ax = murderers_gender_split.plot(kind='bar', color=['blue', 'pink'])

# Add text labels for totals on each bar
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Gender Split of Murderers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.show()
```
